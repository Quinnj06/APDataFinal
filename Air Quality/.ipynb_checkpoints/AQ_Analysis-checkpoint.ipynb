# Which state has the worst average air quality?

```python
group_stage = {"$group": {"_id": "$StateName", "avg_pm25": {"$avg": "$PM25_15yr_Avg" }}}
sort_stage = {"$sort": {"avg_pm25": -1}}
limit_stage = {"$limit": 1}
project_stage = {"$project": {"_id": 0, "StateName": "$_id", "avg_pm25": 1}}

results = collection.aggregate([group_stage, sort_stage, limit_stage, project_stage])

for record in results:
    print(record['StateName'], record['avg_pm25'])
```

Ohio’s average PM2.5 level is relatively high compared to many other states. This is likely driven by the concentration of industrial and manufacturing activity and by dense transportation networks. # What is the average air quality for every region in the US?

(Northeast, Midwest, South, West)

```python
Northeast = [
    "Connecticut",
    "Maine",
    "Massachusetts",
    "New Hampshire",
    "Rhode Island",
    "Vermont",
    "New Jersey",
    "New York",
    "Pennsylvania"
]

Midwest = [
    "Illinois",
    "Indiana",
    "Michigan",
    "Ohio",
    "Wisconsin",
    "Iowa",
    "Kansas",
    "Minnesota",
    "Missouri",
    "Nebraska",
    "North Dakota",
    "South Dakota"
]

South = [
    "Delaware",
    "Florida",
    "Georgia",
    "Maryland",
    "North Carolina",
    "South Carolina",
    "Virginia",
    "West Virginia",
    "Alabama",
    "Kentucky",
    "Mississippi",
    "Tennessee",
    "Arkansas",
    "Louisiana",
    "Oklahoma",
    "Texas"
]

West = [
    "Arizona",
    "Colorado",
    "Idaho",
    "Montana",
    "Nevada",
    "New Mexico",
    "Utah",
    "Wyoming",
    "Alaska",
    "California",
    "Hawaii",
    "Oregon",
    "Washington"
]
```

```python
regions = {"Northeast": Northeast, "Midwest": Midwest, "South": South, "West": West}

region_values = {k: [] for k in regions}

cursor = collection.find({}, {"StateName": 1, "PM25_15yr_Avg": 1, "_id": 0})

for doc in cursor:
    state = doc.get("StateName")
    pm25 = doc.get("PM25_15yr_Avg")
    
    for region, states in regions.items():
        if state in states:
            region_values[region].append(pm25)
            break

region_avg = {r: sum(vals)/len(vals) for r, vals in region_values.items() if vals}


for key in region_avg:
    print(f"{key}: {region_avg[key]}")
```

This is particularly interesting data. Looking at the outlier of the West, which has an average PM2.5 level of approximately 8, one can intuitively assume this is due in part to the region’s much larger geographic size, leading to more spatially dispersed industry and population centers. In contrast, regions like the Northeast and Midwest are more densely populated and industrialized, which can concentrate emissions and result in higher average PM2.5 levels. Additionally, the West’s lower average may mask localized spikes caused by events such as wildfires, which create short-term but intense air pollution that does not fully register in long-term regional averages. Overall, this suggests that population density, industrial clustering, and geographic spread all play a significant role in shaping regional air quality trends.
