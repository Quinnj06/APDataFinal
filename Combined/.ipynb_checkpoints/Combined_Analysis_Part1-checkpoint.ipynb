# How does income affect air quality? 

```python
# vqriable identifying what colum we want to use
income_var = "LowWage_PCT"

# this cuts the df into two halves, lower percentage of low wage workers and a higher percentage of low wage workers.
merged["income_group"] = pd.qcut(merged[income_var], q=2, labels=["Lower % of Low-Wage", "Higher % of Low-Wage"])

merged.groupby("income_group", observed = False)["PM25_15yr_Avg"].mean()
```

Here we see two groups: areas with a lower share of low-wage workers and regions with a higher share. Notably, the average PM2.5 levels between the two groups are very close, suggesting that at this level of aggregation, air quality differences by income group are limited. This implies that while income may influence localized exposure, regional or broad-area averages can mask more nuanced disparities in air pollution exposure.

# How does access to public transport affect air quality?

```python
# resetting the merged df
merged = df_aq.merge(df_walk, on="STC_Fips", how="inner")

merged = merged.drop(columns=["State_Names", "STATEFP_y"])
```

Judging the quality of transit access based on distance from the nearest transit stop

```python
temp = merged[["PM25_15yr_Avg", "Dist_to_TStop_MET"]].copy()

temp["transit_access"] = pd.cut(
    temp["Dist_to_TStop_MET"],
    bins=[0, 200, 500, 5000],
    labels=["Good transit access", "Medium transit access", "Poor transit access"]
)

group = temp.groupby("transit_access", observed = False)["PM25_15yr_Avg"].mean()
display(group)
```

PM2.5 levels are highest in areas with good transit access (12.96), followed by medium (12.71) and poor transit access (12.23). This suggests that higher pollution levels are associated with areas that have better transit access. However, this likely reflects greater urban density, traffic volume, and economic activity in these regions rather than transit access itself being a direct cause of increased air pollution.

# How does car ownership per household affect air quality?

```python
Northeast = [
    "Connecticut",
    "Maine",
    "Massachusetts",
    "New Hampshire",
    "Rhode Island",
    "Vermont",
    "New Jersey",
    "New York",
    "Pennsylvania"
]

Midwest = [
    "Illinois",
    "Indiana",
    "Michigan",
    "Ohio",
    "Wisconsin",
    "Iowa",
    "Kansas",
    "Minnesota",
    "Missouri",
    "Nebraska",
    "North Dakota",
    "South Dakota"
]

South = [
    "Delaware",
    "Florida",
    "Georgia",
    "Maryland",
    "North Carolina",
    "South Carolina",
    "Virginia",
    "West Virginia",
    "Alabama",
    "Kentucky",
    "Mississippi",
    "Tennessee",
    "Arkansas",
    "Louisiana",
    "Oklahoma",
    "Texas"
]

West = [
    "Arizona",
    "Colorado",
    "Idaho",
    "Montana",
    "Nevada",
    "New Mexico",
    "Utah",
    "Wyoming",
    "Alaska",
    "California",
    "Hawaii",
    "Oregon",
    "Washington"
]
```

```python
# creating a region mapping 
region_map = {}

for state in Northeast:
    region_map[state] = "Northeast"

for state in Midwest:
    region_map[state] = "Midwest"

for state in South:
    region_map[state] = "South"

for state in West:
    region_map[state] = "West"
```

```python
# adding specific regions to the merged dataframe
merged["Region"] = merged["StateName"].map(region_map)
display(merged)
```

```python
# grouping by the region and then averaging the car percentages
regional_car_averages = (merged.groupby("Region")[["PM25_15yr_Avg", "0Cars_PCT", "1Cars_PCT", "2pCars_PCT"]].mean())

print(regional_car_averages)
```

This table compares regional car ownership patterns with average PM2.5 levels. Unlike the earlier regional-average analysis, this data shows a different pattern: the West has the lowest PM2.5 despite having the highest share of households with two or more cars. In contrast, the Midwest and Northeast have higher pollution with lower multi-car ownership. This difference highlights a limitation of regional averages: changing the variables being compared can yield different conclusions and mask essential factors such as urban density, industry, and geography.
