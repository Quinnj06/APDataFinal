# How does public transport access affect Walkability 

```python
#find the top public transit access and the lowest and compare their walkability scores
#best transit access
pipeline= [
    {"$sort": {"Ranked_TranDist": -1}},
    {"$limit": 20},
    {"$group":{ "_id" : "null" , "AVG Walkability": {"$avg": "$NatWalkInd"}}}
]

res = walk.aggregate(pipeline)
for r in res:
    print(r)
```

```python
#worst transit access 
pipeline= [
    {"$sort": {"Ranked_TranDist": 1}},
    {"$limit": 20},
    {"$group":{ "_id" : "null" , "AVG Walkability": {"$avg": "$NatWalkInd"}}}
]
res = walk.aggregate(pipeline)
for r in res:
    print(r)
```

To answer this question calculated the average walkability of the 20 areas with the best transit access and 20 areas with the worst transit access. This revealed a drastic change, showing that areas with poor transit access have much worse Walkability, coming in with and average of 8.4 and areas with good public transit access have considerably better Walkability, coming in at 14.8. 

# How does public transport access affect the amount of cars owned per household

```python
#best transit access
pipeline= [
    {"$sort": {"Ranked_TranDist": -1}},
    {"$limit": 20},
    {"$group":{ "_id" : "null" , 
               "Total_without_cars": {"$avg": "$0_CarsOwned"},
               "Total_with_1car":{"$avg": "$1_CarsOwned"},
               "Total_with_2cars":{"$avg": '$2more_CarsOwned'}}}
]
res = walk.aggregate(pipeline)
for r in res:
    print(r)
```

```python
#worst transit access
pipeline= [
    {"$sort": {"Ranked_TranDist": 1}},
    {"$limit": 20},
    {"$group":{ "_id" : "null" , 
               "Total_without_cars": {"$avg": "$0_CarsOwned"},
               "Total_with_1car":{"$avg": "$1_CarsOwned"},
               "Total_with_2cars":{"$avg": '$2more_CarsOwned'}}}
]
res = walk.aggregate(pipeline)
for r in res:
    print(r)
```

For this question I once again found the best and worst 20 areas for transit and then found the average amount of people who owned no cars, one car or two or more cars. We can see through the results that the average amount of people without cars increases in areas with good transit access, and in areas with poor transit access the average amount of people with 2 or more cars also greatly increases. 

# How does road network density affect walkability?

```python
#Highest Road density
pipeline= [
    {"$sort": {'Tot_Road_network_Den': -1}},
    {"$limit": 20},
    {"$group":{ "_id" : "null" , 
              "AVG Walkability": {"$avg": "$NatWalkInd"}}}
]
res = walk.aggregate(pipeline)
for r in res:
    print(r)
```

```python
#Lower road Density 
pipeline= [
    {"$sort": {'Tot_Road_network_Den': 1}},
    {"$limit": 20},
    {"$group":{ "_id" : "null" , 
              "AVG Walkability": {"$avg": "$NatWalkInd"}}}
]
res = walk.aggregate(pipeline)
for r in res:
    print(r)
```

For this question I found the most dense and least dense road systems and calculated the average Walkability. With these results, we can see drastic drop of nearly 11 degrees of walkability between the most dense and least dense areas. This indicates that the road density of an area is a significant factor in the overall walkabilty of an area, even more so than public transit.
